## support data


```python
support_data=pd.read_csv("support_fin_data.csv")
```

```python
support_data.info()
```

```python
## drop time and refernce columns
support_data.columns
```

```python
support_data.drop(columns=['time ','Reference '],inplace=True)
```

```python
support_data.head()
```

```python
support_data.drop(columns=['Previous '],inplace=True)
```

```python
support_data.head()
```

```python
support_data["Actual "]= support_data["Actual "].str.replace('%','')
support_data["Actual "] = support_data["Actual "].str.replace('-','0')
support_data["Actual "] = support_data["Actual "].astype(float)
```

```python
support_data["Consensus"]= support_data["Consensus"].str.replace('%','')
```

```python
support_data.head()
```

```python
### Date into datetime format
support_data['Date'] = pd.to_datetime(support_data['Date'],dayfirst=True)  
```

```python
support_data.head()
```

```python
## match the month and year with fin_data and add actual and consensus columns
support_data['Year'] = support_data['Date'].dt.year
support_data['Month'] = support_data['Date'].dt.month
support_data.head()
```

```python
fin_data.reset_index(inplace=True)
```

```python
### Now we will merge both dataframes on Date column based on month and year of the date 
# 1. Ensure the 'Date' column is in datetime format
fin_data['Date'] = pd.to_datetime(fin_data['Date'])

# 2. Extract the Year and Month into new columns
fin_data['Year'] = fin_data['Date'].dt.year
fin_data['Month'] = fin_data['Date'].dt.month
```

```python
fin_data.head()
```

```python
merged_df = pd.merge(
    fin_data,       # The primary DataFrame (left)
    support_data,   # The data to join (right)
    on=['Year', 'Month'], # The common columns used for matching
    how='left',     # Keep all rows in fin_data and match support_data
    suffixes=('_fin', '_support') # Distinguish columns with the same name (like 'Date')
)
```
