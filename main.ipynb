```python
## convert object to datetime date is in dd /mm/yyyy format

fin_data['Date'] = pd.to_datetime(fin_data['Date'],dayfirst=True)
```

```python
## convert object into integer
## replace , with nothing
fin_data['Price'] = fin_data['Price'].str.replace(',', '')
fin_data['Price'] = fin_data['Price'].astype(float)
```

```python
fin_data["Open"] = fin_data["Open"].str.replace(',', '')
fin_data["Open"] = fin_data["Open"].astype(float)
fin_data["High"] = fin_data["High"].str.replace(',', '') 
fin_data["High"] = fin_data["High"].astype(float)
fin_data["Low"] = fin_data["Low"].str.replace(',', '')
fin_data["Low"] = fin_data["Low"].astype(float)
```

```python
fin_data.head()
```

```python
fin_data.set_index('Date', inplace=True)
```

```python
fin_data["Change %"]= fin_data["Change %"].str.replace('%','')
fin_data["Change %"] = fin_data["Change %"].astype(float)
```

```python
## drop nan values
fin_data.dropna(inplace=True)
```
