```python
def create_sequences(data, lookback_window=10, prediction_horizon=1):
    X, y = [], []
    for i in range(len(data) - lookback_window - prediction_horizon + 1):
        X.append(data[i:(i + lookback_window)])
        y.append(data[i + lookback_window : i + lookback_window + prediction_horizon, PRICE_COLUMN_INDEX]) 
        
    return np.array(X), np.array(y)
```

```python
def build_attention_lstm_model(timesteps, features, output_dim):
    # Input Layer
    inputs = Input(shape=(timesteps, features))
    
    # Encoder: LSTM Layer
    # return_sequences=True is CRITICAL to provide all hidden states (Keys/Values)
    lstm_out, h_state, c_state = LSTM(
        units=64, 
        return_sequences=True, 
        return_state=True  # h_state is the final hidden state
    )(inputs)

    # --- Attention Mechanism FIX ---
    # Keras Attention layer expects the inputs to be (query, value) or (query, value, key)
    # The 'value' (lstm_out) is 3D: (samples, timesteps, units)
    # The 'query' (h_state) is 2D: (samples, units) and needs to be reshaped to 3D (samples, 1, units)
    # so that it can be broadcast for the attention operation.
    
    # Reshape h_state to (None, 1, 64) to be compatible with the Attention layer's value input
    query_tensor = keras.layers.Reshape((1, 64))(h_state)

    # Apply Attention. The output (attention_out) will be 3D: (samples, 1, units).
    attention_output = Attention(use_scale=True)([query_tensor, lstm_out])
    
    # Flatten the attention output from (None, 1, 64) to (None, 64)
    # This ensures both tensors for concatenation are 2D.
    attention_flat = keras.layers.Flatten()(attention_output)

    # --- Concatenation FIX ---
    # lstm_out[:, -1, :] already extracts the last timestep and is 2D (None, 64)
    # We now concatenate the last step of the LSTM output with the flattened context vector
    concat_layer = keras.layers.Concatenate(axis=-1)([lstm_out[:, -1, :], attention_flat])

    # --- Decoder/Output Layers ---
    dense_1 = Dense(32, activation='relu')(concat_layer)
    dropout = Dropout(0.2)(dense_1)
    
    # Output Layer
    outputs = Dense(output_dim, activation='linear')(dropout)
    
    # Create Model
    model = Model(inputs=inputs, outputs=outputs)
    
    return model

# Build the model using the fixed function
model = build_attention_lstm_model(N_TIMESTEPS, N_FEATURES, N_OUTPUT)

# Compile and Train (as before)
model.compile(optimizer='adam', loss='mse', metrics=['mae']) 
model.summary()
```
